```python
original_data.columns
```

```python
cols_with_missing = [col for col in original_data.columns 
                                 if original_data[col].isnull().any()]
print(cols_with_missing)
```

```python
data_not_missing_value = original_data.fillna(original_data['Price'].mean())
data_not_missing_value
```

```python
data_not_missing_value.info()
```

```python
data_not_missing_value1 = data_not_missing_value.groupby(['Suburb'])['Price'].agg(np.sum).reset_index()
data_not_missing_value1
```

```python
f, ax = plt.subplots(figsize = (25,9))
data_not_missing_value1.sort_values(by = ['Price'], inplace = True, ascending = False)
sns.barplot(x = 'Price', y = 'Suburb', data = data_not_missing_value1.head(10))
ax.set_title('Top 10 Suburb by the Price', fontsize = 20)
ax.set_xlabel('Price', fontsize = 15)
ax.set_ylabel('Suburb', fontsize = 15)
plt.show()
```

```python
features = ['Rooms','Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'Price' ]
```

```python
data = data_not_missing_value[features]
```
